```
from urllib import request # for downloading data from url for web scrapping
from bs4 import BeautifulSoup # for souping
import time
import re # importing regular expression
#pointing to url
url='https://en.wikipedia.org/wiki/Machine_learning'
#url1='https://www.google.com/covid19/'

htmldata=request.urlopen(url)
#htmldata.read() # it will download data in html formate
soupdata=BeautifulSoup(htmldata,'html5lib')# html data parser

atagdata=soupdata.findAll('p')
#print(atagdata)
#now converting data into string from HTML format
mydata=""
for i in atagdata:
  mydata += i.text
#print(mydata)
# data cleaning
clean_data=re.sub(r'\[[0-9]*\]',' ',mydata) #this will remove 0 or more time number apearing in my data
clean_data=re.sub(r'\s+',' ',clean_data) # it will remove one or more white space with one whitespace
clean_data=re.sub(r'[^a-zA-Z]',' ',clean_data) # it will remove single char from starting of line
clean_data=re.sub(r'\s+',' ',clean_data) # it will remove one or more white space with one whitespace
print(clean_data)
print(type(clean_data))
newdata=clean_data.split()
#deal with data by using NLTK
import nltk#downloading all dictionary for computer
#nltk.download('all')
#first removing the stop words
mypuredata=[]
from nltk.corpus import stopwords
for i in newdata:
  if i.lower() not in stopwords.words('english'):
    mypuredata.append(i)
print(mypuredata)
freq_data=nltk.FreqDist(newdata)
freq_data.plot(20) #top 20 words having the most count    
```
